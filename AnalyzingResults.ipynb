```python
# Importing the necessary libraries
import os
import pandas as pd
import matplotlib.pyplot as plt

# Define the base path where the "model_results" folders are located
base_path = "./"

# List to store all results found
all_results = []

# Function to search for result files in "model_results" folders
def search_results(base_path):
    for root, _, files in os.walk(base_path):
        if "model_results" in root and "training_results.csv" in files:
            csv_path = os.path.join(root, "training_results.csv")
            print(f"Found: {csv_path}")
            try:
                df = pd.read_csv(csv_path)
                if {'model', 'column', 'search_type', 'r2_score', 'rmse', 'params', 'run'}.issubset(df.columns):
                    all_results.append(df)
                else:
                    print(f"File {csv_path} ignored due to missing columns.")
            except Exception as e:
                print(f"Error reading file {csv_path}: {e}")

# Call the search function
search_results(base_path)

# Combine all results into a single DataFrame
if all_results:
    combined_results = pd.concat(all_results, ignore_index=True)

    # Saving the top 5 configurations for each model and column
    best_5_per_model = (
        combined_results
        .sort_values(by=['column', 'model', 'r2_score'], ascending=[True, True, False])
        .groupby(['model', 'column'])
        .head(5)  # Selects the top 5 for each model and column
    )
    best_5_per_model.to_csv("best_5_per_model.csv", index=False)
    print("File 'best_5_per_model.csv' saved successfully.")
    
    # Finding the best overall model for each column
    best_model_per_column = combined_results.loc[
        combined_results.groupby(['column'])['r2_score'].idxmax()
    ]
    best_model_per_column.to_csv("best_model_per_column.csv", index=False)
    print("File 'best_model_per_column.csv' saved successfully.")

    # Detailed analysis with tables and graphs
    for column in combined_results['column'].unique():
        column_data = combined_results[combined_results['column'] == column]
        
        # Table: Top 5 best results by model
        print(f"\nTop 5 best results by model for column '{column}':")
        top_5_table = best_5_per_model[best_5_per_model['column'] == column]
        display(top_5_table)
        
        # Table: Best overall model for the column
        print(f"\nBest overall model for column '{column}':")
        best_model_table = best_model_per_column[best_model_per_column['column'] == column]
        display(best_model_table)
        
        # Graph: Top 5 scores by model
        plt.figure(figsize=(12, 8))
        for model in column_data['model'].unique():
            model_data = column_data[column_data['model'] == model].nlargest(5, 'r2_score')
            plt.plot(
                range(len(model_data)), 
                model_data['r2_score'], 
                marker='o', 
                label=model
            )
            for i, score in enumerate(model_data['r2_score']):
                plt.text(i, score, f"{score:.2f}", fontsize=8)
        
        plt.title(f"Top 5 Scores by Model - Column {column}")
        plt.ylabel("R2 Score")
        plt.xlabel("Configurations")
        plt.legend(title="Models", loc='best')
        plt.grid()
        plt.tight_layout()
        plt.savefig(f"top_5_scores_{column}.png")  # Saves the graph as an image
        plt.show()

else:
    print("No 'training_results.csv' files found in 'model_results' folders.")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the base path where the "model_results" folders are located
base_path = "./"

# List to store all results found
all_results = []

# Function to search for result files in "model_results" folders
def search_results(base_path):
    for root, _, files in os.walk(base_path):
        if "model_results" in root and "training_results.csv" in files:
            csv_path = os.path.join(root, "training_results.csv")
            print(f"Found: {csv_path}")
            try:
                df = pd.read_csv(csv_path)
                # Ensure the required columns are present in the file
                if {'model', 'column', 'search_type', 'r2_score', 'rmse', 'params', 'run'}.issubset(df.columns):
                    all_results.append(df)
                else:
                    print(f"File {csv_path} skipped due to missing columns.")
            except Exception as e:
                print(f"Error reading file {csv_path}: {e}")

# Call the function to search for result files
search_results(base_path)

if all_results:
    # Combine all results into a single DataFrame
    combined_results = pd.concat(all_results, ignore_index=True)

    # Rename 'column' values to 'M0, M1, ...'
    unique_columns = combined_results['column'].unique()
    column_mapping = {col: f"M{i}" for i, col in enumerate(unique_columns)}
    combined_results['column'] = combined_results['column'].map(column_mapping)

    # Filter out rows with negative R2 scores
    combined_results = combined_results[combined_results['r2_score'] >= 0]

    # Analyze each column separately
    for column in combined_results['column'].unique():
        column_data = combined_results[combined_results['column'] == column]

        # Calculate lower bound for pruning based on IQR
        q1 = column_data['r2_score'].quantile(0.25)
        iqr = column_data['r2_score'].quantile(0.75) - q1
        lower_bound = q1 - 1.5 * iqr
        pruned_data = column_data[column_data['r2_score'] >= lower_bound]

        # Scatter plot for R2 scores by run
        plt.figure(figsize=(12, 6))
        sns.scatterplot(
            data=pruned_data,
            x="run",
            y="r2_score",
            hue="model",
            style="search_type",
            s=100,
            alpha=0.7
        )
        plt.title(f"R2 Score Scatter Plot by Run - Column '{column}'")
        plt.xlabel("Run")
        plt.ylabel("R2 Score")
        plt.legend(title="Models")
        plt.grid()
        plt.tight_layout()
        plt.savefig(f"scatter_r2_pruned_{column}.png")
        plt.show()

        # Boxplot for R2 score distribution
        plt.figure(figsize=(12, 6))
        sns.boxplot(
            data=pruned_data,
            x="model",
            y="r2_score",
            hue="search_type"
        )
        plt.title(f"R2 Score Distribution (Lower Pruned) by Model - Column '{column}'")
        plt.xlabel("Models")
        plt.ylabel("R2 Score")
        plt.xticks(rotation=45)
        plt.legend(title="Search Type", loc='best')
        plt.grid()
        plt.tight_layout()
        plt.savefig(f"boxplot_r2_pruned_{column}.png")
        plt.show()

    # Find the best model for each column based on R2 score
    best_model_per_column = combined_results.loc[
        combined_results.groupby(['column'])['r2_score'].idxmax()
    ]

else:
    print("No 'training_results.csv' files found in 'model_results' folders.")
```

```python
# Importing required libraries
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the base path where "model_results" folders are located
base_path = "./"

# List to store all the results found
all_results = []

# Function to search for result files in "model_results" folders
def search_results(base_path):
    for root, _, files in os.walk(base_path):
        if "model_results" in root and "training_results.csv" in files:
            csv_path = os.path.join(root, "training_results.csv")
            print(f"Found: {csv_path}")
            try:
                df = pd.read_csv(csv_path)
                # Check if the required columns are present
                if {'model', 'column', 'search_type', 'r2_score', 'rmse', 'params', 'run'}.issubset(df.columns):
                    all_results.append(df)
                else:
                    print(f"File {csv_path} skipped due to missing columns.")
            except Exception as e:
                print(f"Error reading file {csv_path}: {e}")

# Call the function to search for result files
search_results(base_path)

# Combine all results into a single DataFrame
if all_results:
    combined_results = pd.concat(all_results, ignore_index=True)

    # Rename 'column' values to 'M0, M1, ...' for better visualization
    unique_columns = sorted(combined_results['column'].unique())
    column_mapping = {col: f"M{i}" for i, col in enumerate(unique_columns)}
    combined_results['column'] = combined_results['column'].map(column_mapping)

    # Remove rows with r2_score less than 0
    combined_results = combined_results[combined_results['r2_score'] >= 0]

    # Identify the best model configuration for each column
    best_model_per_column = combined_results.loc[
        combined_results.groupby(['column'])['r2_score'].idxmax()
    ]

    # Fix the ordering of columns for display
    best_model_per_column['column_order'] = best_model_per_column['column'].str.extract(r'M(\d+)').astype(int)
    best_model_per_column = best_model_per_column.sort_values('column_order').drop(columns=['column_order'])

    print("\nBest model configuration for each column:")
    display(best_model_per_column)

    # Plot: Comparison of best models for each column
    plt.figure(figsize=(12, 6))
    sns.barplot(
        data=best_model_per_column,
        x="column",
        y="r2_score",
        hue="model"
    )
    plt.title("Best Model by Column (Based on R2 Score)")
    plt.xlabel("Columns")
    plt.ylabel("R2 Score")
    plt.xticks(rotation=45)
    plt.legend(title="Models")
    plt.grid()
    plt.tight_layout()
    plt.savefig("best_model_per_column_sorted.png")
    plt.show()

    # Find the best overall result for each column
    best_results = combined_results.groupby("column")["r2_score"].max().reset_index()

    # Fix the ordering of columns
    best_results['column_order'] = best_results['column'].str.extract(r'M(\d+)').astype(int)
    best_results = best_results.sort_values('column_order').drop(columns=['column_order'])

    print("\nBest overall results (based on R2) for each column:")
    display(best_results)

    # Plot: Best results for each column
    plt.figure(figsize=(12, 6))
    sns.barplot(
        data=best_results,
        x="column",
        y="r2_score",
        palette="viridis"
    )
    plt.title("Best Results by Column (Based on R2 Score)")
    plt.xlabel("Columns")
    plt.ylabel("R2 Score")
    plt.xticks(rotation=45)
    plt.grid()
    plt.tight_layout()
    plt.savefig("best_results_per_column_sorted.png")
    plt.show()

else:
    print("No 'training_results.csv' files found in 'model_results' folders.")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the base path where the "model_results" folders are located
base_path = "./"

# List to store all results
all_results = []

# Function to search for "training_results.csv" files in "model_results" folders
def buscar_resultados(base_path):
    for root, _, files in os.walk(base_path):
        if "model_results" in root and "training_results.csv" in files:
            csv_path = os.path.join(root, "training_results.csv")
            print(f"Found: {csv_path}")
            try:
                df = pd.read_csv(csv_path)
                if {'model', 'column', 'r2_score'}.issubset(df.columns):
                    all_results.append(df)
                else:
                    print(f"File {csv_path} skipped due to missing columns.")
            except Exception as e:
                print(f"Error reading file {csv_path}: {e}")

# Call the function to search for results
buscar_resultados(base_path)

# Combine all results into a single DataFrame
if all_results:
    combined_results = pd.concat(all_results, ignore_index=True)

    # Filter out TabNet results
    combined_results = combined_results[combined_results['model'] != 'TabNet']

    # Rename columns for clarity (e.g., M0, M1, ...)
    unique_columns = sorted(combined_results['column'].unique())
    column_mapping = {col: f"M{i}" for i, col in enumerate(unique_columns)}
    combined_results['column'] = combined_results['column'].map(column_mapping)

    # Remove invalid R scores
    combined_results = combined_results[combined_results['r2_score'] >= 0]

    # Scatter plot
    plt.figure(figsize=(12, 6))
    sns.scatterplot(
        data=combined_results,
        x="column",
        y="r2_score",
        hue="model",
        style="model",
        alpha=0.7,
        s=100
    )
    plt.title("Scatter Plot of R Scores by Column and Model")
    plt.xlabel("Columns")
    plt.ylabel("R Score")
    plt.xticks(rotation=45)
    plt.legend(title="Models", loc='best')
    plt.grid()
    plt.tight_layout()
    plt.savefig("scatter_r2_by_column_and_model_no_tabnet.png")
    plt.show()

else:
    print("No 'training_results.csv' files found in 'model_results' folders.")
```
